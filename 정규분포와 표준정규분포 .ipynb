# **표준 정규 분포** 

표준 정규 분포의 확률밀도함수는 다음과 같다.

$f(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{z^2}{2}}$

$\frac{1}{\sqrt{2\pi}}$ 는 표준 정규 분포의 적분값이 1이 되도록 정규화 해주는 정규화 상수이다. 

표준 정규 분포는 여러 분포 중 가장 많이 볼 수 있는 분포로서 통계학에 대해서 잘 알지 못하더라도 한 번씩은 들어봤을  분포이다.

```python
X = np.linspace(-3,3,100)
pdf = 1/np.sqrt(2*np.pi) * np.exp(-X**2/2)


fig,ax = plt.subplots()

ax.plot(X,pdf, c = 'blue')
ax.set_xlabel('확률변수')
ax.set_ylabel('확률밀도')
ax.set_title('표준 정규 분포')
```

# **표준정규분포의 기대값과 분산**

**기대값**

$E(X) = \int_{-\infty}^{+\infty} x f(x) dx$ 

$E(X) = \int_{-\infty}^{+\infty} x \frac{1}{\sqrt{2\pi}} e^{-\frac{z^2}{2}} dx$

인데 표준 정규분포는 평균을 대칭으로 하는 기함수임을 알 수 있다. 

그럼으로 적분값의 합은 0이 된다.

$E(X) = 0$

**분산**

$Var(x) = E(X^2) - E(X)^2$

# **표준 정규 분포의 누적 확률 분포**


$\Phi(Z) = \frac{1}{\sqrt{\frac{1}{2\pi}}}\int_{-\infty}^Z e^{-\frac{z^2}{2}}$

로 나타낼 수 있는데 

매 번 식을 작성 할 수 없으니 대부분 컴퓨터엔 표준 정규 분포의 Z 값에 대한 누적 확률 밀도 함수 값들이 계산 되어 있다.

# **정규 분포**

정규분포는 

$N$~$(\mu,\sigma^2)$

로 표기한다.

이 때 정규분포를 따르는 확률 분포의 X는 다음을 따른다.

$X$ = $\mu  + \sigma * Z$ 

여기서 $Z$ 는 $X$를 표준정규분포 위의 확률 변수 X로서, 정규화 된 확률변수 X를 의미한다.

해당 식의 의미는 

정규분포를 따르는 어떤 확률 변수 X는 

표준 정규분포를 $\sigma$ 만큼 늘리고, 평균의 위치를 0에서 $\mu$로 수평이동 시킨 값을 의미한다.

예시를 보기 위해 $N$~$(10,5^2)$ 를 따르는 정규 분포가 있다고 해보자 

```python
X = np.linspace(-3,3,100)
Z = 1/np.sqrt(2*np.pi) * np.exp(-X**2/2)

mu,sigma = 10,5

fig,ax = plt.subplots(ncols = 3,figsize = (20,5))

ax[0].plot(X,Z, c = 'blue')
ax[0].set_xlabel('확률변수')
ax[0].set_ylabel('확률밀도')
ax[0].set_title('표준 정규 분포')


ax[1].plot(X * sigma, Z,c = 'blue')
ax[1].set_xlabel('확률변수')
ax[1].set_title('표준 정규 분포를 $\sigma$만큼 늘림')

ax[2].plot(X * sigma + mu, Z,c = 'blue')
ax[2].set_xlabel('확률변수')
ax[2].set_title('표준 정규 분포를 $\sigma$만큼 늘린 후 $\mu$만큼 수평이동')

fig.suptitle('$N$~$(10,5^2)$의 그래프 \n 표준 정규 분포로부터의 변화 과정', y = 1.1)
```

# **정규화 Z**

정규분포를  따르는 확률 변수 X 는 

$X = \mu + \sigma*Z$

라고 했다. 해당 식을 Z에 대해 풀어보면

$Z = \frac{X - \mu}{\sigma}$

가 나온다. 

우리가 데이터를 정규화 시켜줄 때 사용했던 해당 식이, 데이터가 표준 정규 분포로부터 시작되어 만들어진 정규분포를 따르기 때문에 사용 가능했던 것이다 !  

# **정규분포의 기대값과 분산**

$X = \mu + \sigma*Z$

기에 기대값의 선형성을 이용해서 구해주면

$E(X) = E(\mu) + E(\sigma*Z)$

$= \mu + \sigma *E(Z)$

= $\mu$

가 나온다. 

----


분산은

$Var(X) = Var(\mu + \sigma*Z)$

$= Var(\mu) + \sigma^2  * Var(Z)$

$= 0 + \sigma^2 * 1 $

$= \sigma ^2$

# **정규 분포의 확률밀도함수, 누적확률밀도함수**

확률밀도함수는 누적확률밀도함수의 도함수이다.

이에 누적 확률 밀도 함수부터  구해보자

**$X$의 $CDF$**

----


$P(X \le x)$ 

를 양 변을 정규화 시켜주자 

$P(\frac{X -\mu}{\sigma} \le \frac{z - \mu}{\sigma})$

$= P(Z \le  z) $

$= \Phi(Z)$

정규분포의 누적확률밀도함수는 정규화 시킨 후 표준정규분포 위에서의 누적확률밀도함수 값과 같다.

----

**$X$의 $PDF$**

PDF 는 CDF 의 미분값이다.

$\Phi(\frac{x - \mu}{\sigma})$

을 x 에 대해  미분하면 연쇄법칙에 의해

$\frac{1}{\sigma} *  \Phi \frac{d\Phi}{dx}$

이 때 표준  정규 분포의 누적확률밀도함수를 미분한 것은 표준 정규 분포의 확률밀도함수가 되니

$f(x) = \frac{1}{\sqrt{2\pi}  \sigma} e^{-\frac{(\frac{x - \mu}{\sigma})^2}{2}}$

가 된다.

# **68 - 95 - 99.7 % Rule**

$P(|X -  \sigma| \le \mu ) \approx 0.68$

$P(|X -  2*\sigma| \le \mu ) \approx 0.95$

$P(|X -  3*\sigma| \le \mu ) \approx 0.997$

정규분포를 따르는 데이터가 

각 평균으로부터 1,2,3 표준편차 떨어진 곳 사이에 위치할 확률이 각 0.68,0.95,0.997  이 된다.  

```python
X  =  np.linspace(-3, 3, 100)
pdf  =  1 / np.sqrt(2 * np.pi) * np.exp(-X**2 / 2)

fig, ax  =  plt.subplots()

ax.plot(X, pdf, c = 'blue')

ax.fill_between(X, pdf, where = (-1 <=  X) & (X <=  1), color = 'gray', alpha = 0.3, label = '68%')
ax.fill_between(X, pdf, where = (-2 <=  X) & (X <=  2), color = 'orange', alpha = 0.3, label = '95%')
ax.fill_between(X, pdf, where = (-3 <=  X) & (X <=  3), color = 'green', alpha = 0.3, label = '99.7%')

ax.set_xlabel('확률변수')
ax.set_ylabel('확률밀도')
ax.set_title('표준 정규 분포')

ax.legend()

# Annotating the ranges
ax.annotate('$\mu - \sigma$', xy = (-1, 0), xytext = (-1.5, 0.1), arrowprops = dict(arrowstyle = '->'), fontsize = 10)
ax.annotate('$\mu + \sigma$', xy = (1, 0), xytext = (1.5, 0.1), arrowprops = dict(arrowstyle = '->'), fontsize = 10)

ax.annotate('$\mu - 2\sigma$', xy = (-2, 0), xytext = (-2.5, 0.03), arrowprops = dict(arrowstyle = '->'), fontsize = 10)
ax.annotate('$\mu + 2\sigma$', xy = (2, 0), xytext = (2.5, 0.03), arrowprops = dict(arrowstyle = '->'), fontsize = 10)

ax.annotate('$\mu - 3\sigma$', xy = (-3, 0), xytext = (-3.5, 0.01), arrowprops = dict(arrowstyle = '->'), fontsize = 10)
ax.annotate('$\mu + 3\sigma$', xy = (3, 0), xytext = (3.5, 0.01), arrowprops = dict(arrowstyle = '->'), fontsize = 10)

plt.show()
```
