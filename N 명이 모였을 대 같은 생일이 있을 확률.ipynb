### __생일 문제__ 

* 몇 명의 사람이 모여있을 때 같은생일인 사람이 있을 확률이 50%를 넘을까 ? 

```python
def birth_matching(n):
    '''
    n 이 365 이상일 경우엔 필연적으로 생일이 같은 사람이 존재한다
    n 이 365 이하 일 때 생일이 모두 다를 확률을 구해보자 
    '''
    
    total_event = 365 ** n # n 명의 인원이 가질 수 있는 생일의 경우의 수 
    birth_perm = math.perm(365,n) # 365일의 생일들에 n 을 집어넣는 경우의 수 
    
    # 생일이 모두 다를 경우는 
    # 첫 번째 사람이 365 개 중 하나, 두 번 째 사람이 364 개 중 하나 , 세 번 째 사람이 363개 중 하나 .. 
    
    # 365 * 364 * ... * 365 - n + 1 즉 365 P n 이다.
    
    not_matcing = birth_perm / total_event
    
    return 1 - not_matcing
```

```python
percent = []

for i in range(365):
    percent.append(birth_matching(i) * 100)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

fig,ax = plt.subplots(figsize = (10,5))

ax.plot(percent)
ax.set_xlabel('Number of person')
ax.set_ylabel('Probability')
ax.set_title('Birthday Matching')
```

**50명만 넘어가도 같은 생일인 사람이 있을 확률이 100% 에 수렴한다**


**25명만 넘어가도 같은 생일인 사람이 있을 확률이 50%가 넘는다**
