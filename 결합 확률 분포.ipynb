
### **결합 확률 분포**


여태 까지의 강의에서는 하나의 확률 변수의 분포에 대한 내용들을 주로 다뤘는데


이번에는 두 가지 이상의 결합 확률 분포에 대해서 공부해봤다.


결합 확률 분포는 특정 조합에서 함께 발생하는 여러 확률 변수의 확률을 제공하는 확률 분포이다.


두 개 이상의 무작위 변수가 관련된 공동 사건의 확률을 특성화 하며 f(X,Y) 와 같은 식으로 표기한다.

﻿
결합 확률 분포도 하나의 확률 분포이기 때문의 모든 확률의 합은 1이 되며, 각 확률은 0 이상의 값을 가져야 하는 조건을 갖는다.


그럼 만약 결합 확률 분포가 아니라, X , Y 에 대해서만 구하고 싶다면 어떻게 해야 할까 ?


결합 확률 분포에서 결합된 형태가 아닌 부분적인 부분들의 확률 분포를 Marginal Distribution 이라고 한다.


X 에 대한 Marginal distribution 을 구해보자.


Marginal distribution 은 전체 확률의 법칙에 의해 구할 수 있다.

﻿

$Let\ X\ \sim \ Uniform\left(0,5\right)\ ,\ Y\ \sim \ Uniform\left(0,1\right)$

$f_x\left(x\right)=\int _0^1f\left(x,y\right)dy$

$f_y\left(y\right)=\int _0^{5}f\left(x,y\right)dx$

```python

x_min, x_max = 0, 5
y_min, y_max = 0, 2

rectangle = plt.Rectangle((x_min, y_min), x_max - x_min, y_max - y_min, linewidth=1, edgecolor='None', facecolor='dodgerblue')

plt.figure()
ax = plt.gca()
ax.add_patch(rectangle)

# Set the x and y axis limits
plt.xlim(0, 6)
plt.ylim(0, 3)

# Set axis labels
plt.xlabel('X')
plt.ylabel('Y')

for i in range(6):
    plt.plot([i,i],[0,2], c = 'w', ls = ':')
    
plt.plot([0,5],[1,1], c = 'w', ls = ':')

for i in range(1,6):
    plt.text(i- 0.7, 1.5, s = '1/10', c = 'w')
    plt.text(i- 0.7, 0.5, s = '1/10', c = 'w')
    
    
plt.yticks([1,2,3])

plt.title('$f(X,Y)$ \n $X~Uniform(0,5), Y ~ Uniform(0,2)$')
```
