### **베르누이 시행**

```python
p = 0.5
q = 1 - 0.5

fig,ax = plt.subplots(figsize = (5,3))

sns.barplot(x = [0,1],
            y = [p,q],
            ax = ax)

ax.set_title('베르누이 분포')
ax.set_xticklabels(['앞면','뒷면'])
ax.set_yticklabels([str(round(_,2)) + '%' for _ in ax.get_yticks()]);
```


베르누이 시행이란 (영어 강좌에선 대부분 버눌리라고 발음하더라) 가능한 사건이 두 개밖에 없는 시행을 의미한다.


각 사건을 성공과 실패로 정의하였을 때 성공할 확률이 p 라면 실패할 확률은 1-p 를 갖는다.


전체 표본 공간의 모든 사건의 확률의 합은 1이 되어야 하기 때문이다.


적절한 예로는 동전을 던졌을 때 앞면이 나오거나, 뒷면이 나오는 사건이나


혹은 게임 아이템을 강화할 때 성공할 확률이 40% 인 시행등을 의미한다. [이 때 실패할 확률은 1 - 0.4 로 0.6 이 될 것이다.]

### **이항 분포**


이항 분포는 베르누이 시행을 i.i.d 하게 n 번 시행했을 때 X 번 성공하는 사건의 분포를 의미한다.


즉 내가 동전을 n 번 던졌을 때 가능한 성공 횟수는 0번부터 n 번까지 존재하는데


각 사건에 대한 분포를 이항 분포라고 한다


이 때 각 0번부터 n 번까지의 시행은 i.i.d 해야 한다.

﻿

```python
p = 0.5 

probs = []

for _ in range(1000):
    success = 0
    for  i in range(100):
        if np.random.randn() > p:
            success += 1 
            
    probs.append(success)
```

```python
fig,ax = plt.subplots(figsize = (10,5))

sns.histplot(data = probs,bins = 100,ax  = ax)
ax.set_title('동전 던지기의 이항 분포')
ax.set_xlabel('성공 횟수')
```


### **이항 분포의 확률 질량 함수**


$B\left(n,p\right)\ =\ \binom{n}{k}p^k\ \left(1-p\right)^{n\ -k}$


수식에서 p^k (1-p)^ n-k 는 n 번 시도하여 k 번 성공할 확률에 대한 확률 값을 나타내고

​

nCk 부분은 n  번의 시행 중 k개의 성공이 위치할 경우의 수값을 의미한다.

​

표기법은 다음과 같다.

$X $~$B(n,p)$

성공 확률이 p 인 베르누이 시행을 n 번 시행하였을 때 성공한 사건 X 에 대한 분포를 다음과 같의 표기한다.

​

이항 분포는 베르누이 시행을 i.i.d 하게 n 번 시행했을 때 X 번 성공하는 사건의 분포를 의미한다.

​

즉 내가 동전을 n 번 던졌을 때 가능한 성공 횟수 k는 0번부터 n 번까지 존재하는데

​

 각 사건에 대한 분포를 이항 분포라고 한다

​

 이 때 각 0번부터 n 번까지의 시행은

 i.i.d 해야 한다.


```python
def binomial_distribution(p,n):
    
    fig,ax = plt.subplots(figsize = (10,5))
    
    probs = []
    
    for k in range(n + 1):
        
        success = p ** k 
        failuer = (1-p) ** (n-k)
        
        probs.append(math.comb(n,k)  * success * failuer)
        
    ax.plot(probs)
    ax.set_title(f'X ~ B({n},{p})')
    ax.set_xlabel('성공 횟수')        
    
```

```python
binomial_distribution(0.5,100)
```
